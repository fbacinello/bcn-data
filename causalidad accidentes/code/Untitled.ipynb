```python
import folium

#from folium.plugins import HeatMap, MarkerCluster
```

```python
# Create a map
m_1 = folium.Map(location=[41.389920, 2.184503], tiles='openstreetmap', zoom_start=12)
m_2 = folium.Map(location=[41.389920, 2.184503], tiles='cartodbpositron', zoom_start=12)

# Display the map
m_1
```

```python
import geopandas as gpd
import pandas as pd

dataframe = pd.read_csv('../data/2019_accidents_causes_gu_bcn_.csv')
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe['Descripcio_causa_mediata'].value_counts()
```

```python
exceso_velocidad = dataframe.loc[dataframe['Descripcio_causa_mediata'] == 'Excés de velocitat o inadequada']
```

```python
from folium import Marker
# Add points to the map
for idx, row in exceso_velocidad.iterrows():
    Marker([row['Latitud'], row['Longitud']]).add_to(m_2)
m_2
```

```python
alcoholemia = dataframe.loc[dataframe['Descripcio_causa_mediata'] == 'Alcoholèmia']
# Add points to the map
for idx, row in alcoholemia.iterrows():
    Marker([row['Latitud'], row['Longitud']]).add_to(m_2)
m_2
```

```python
from folium.plugins import HeatMap

# Create a base map
m_3 = folium.Map(location=[41.389920, 2.184503], tiles='cartodbpositron', zoom_start=12)

# Add a heatmap to the base map
HeatMap(data=alcoholemia[['Latitud', 'Longitud']], radius=10).add_to(m_3)

# Display the map
m_3
```
